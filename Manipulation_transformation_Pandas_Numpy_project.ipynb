**Max**

````python
max_b = ds_movies.budget.max()
print(f"This isn the max budget on a movie: {max_b}")

#We have to show the number in the best way

max_b_formatted = "{:,.2f}".format(max_b)

#We replace the , for .
max_b_formatted = max_b_formatted.replace(',', '.')

#We use the format ",.2", means we want separators of thousands and two decimal places
print("---")
print(f"This is the max budget on a movies: ${max_b_formatted}")
````

#Conditions

**New DataFrame**

````python
#We are use the columns: budget and title and the revenue  theses examples.

title = ds_movies["title"]
budget = ds_movies["budget"]
revenue = ds_movies["revenue"]

#create the new DataFrame with both columns
df_t_b = pd.DataFrame({"title": title, "budget": budget, "revenue": revenue})

print(df_t_b)
````

**Smaller than**

````python
#smaller than
#print("budget < revenue")

print("Movies with more budget than revenue")
#create the condition with the columns
condition = df_t_b["budget"] < df_t_b["revenue"]

#Filtered to see the titles
filtered_titles = df_t_b.loc[condition, "title"]

print(filtered_titles)
````

**Revenue more than budget**

````python
#Revenue more than budget
#print("budget > revenue")

print("Movies with more revenue than budget")
#create the condition with the columns
condition = df_t_b["budget"] > df_t_b["revenue"]

#Filtered to see the titles
filtered_titles = df_t_b.loc[condition, "title"]

print(filtered_titles)
````
